## **Test for Increase in Sales**

 **Null Hypothesis (H0)**: There is no increase in sales after stepping into digital marketing.

**Alternative Hypothesis (H1)**: There is a significant increase in sales after stepping into digital marketing.

## **Paired-sample t-test**

```python
import pandas as pd
from scipy.stats import ttest_rel
```

```python
t_stat, p_value = ttest_rel(df['Sales_before_digital_add(in $)'],df['Sales_After_digital_add(in $)'])
```

```python
t_stat
```

```python
p_value
```

```python
if p_value < 0.05:
  print('Reject the null hypothesis')
else:
  print('Accept the null hypothesis')
```

## **Check for Dependency between Region and Manager**

**Null Hypothesis (H0)**: There is no significant dependency between Region and Manager.

**Alternative Hypothesis (H1)**: There is a significant dependency between Region and Manager.

## **Chi-square test for independence**

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(df['Region'],df['Manager'])
```

```python
contingency_table
```

```python
chi2_contingency(contingency_table)
```

```python
if p_value < 0.05:
  print('Reject null hypothesis')
else:
  print('Accept null hypothesis')
```
